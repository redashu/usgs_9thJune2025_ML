```python
# importing python libs 
import pandas as pd 
from sklearn.impute import SimpleImputer # fixing missing values 
import numpy as np
from sklearn.preprocessing import LabelEncoder # label encoding for 2 values in single column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
# calling DecisionTree classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score , confusion_matrix
import matplotlib.pyplot as plt 
from sklearn.tree import plot_tree
```

```python
# github raw source
df2 = pd.read_csv('https://raw.githubusercontent.com/redashu/Datasets/refs/heads/master/data_preprocess1.csv')
```

```python
# seperating data into input and output 
# selecting required features and labels

input_features = df2.iloc[:,0:3].values
label = df2.iloc[:,-1].values
```

```python
# printing label as well
label
```

```python
# fixing age missing value 
ashu_impute = SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
ashu_impute.fit(input_features[:,1:3]) # calculating the missing values 
input_features[:,1:3] = ashu_impute.transform(input_features[:,1:3]) # making changes
```

```python
print(input_features)
```

```python
# so finally features and label 
label
```

```python
# converting 2 values into 0 and 1ones 
le = LabelEncoder()
label = le.fit_transform(label)
print(label)
```

```python
# data preprocessing using categorical values 
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
input_features=np.array(ct.fit_transform(input_features))
print(input_features)
```

```python
# finally we have input and label this one we can train to Machine Learning ALgo
traing_input,testing_input,train_label,test_label=train_test_split(input_features,label,test_size=0.1)
# note test_size (0 to 1) --- 0.1 means 10% testing data size
```

```python
# calling DecisionTree classifier
classifier = DecisionTreeClassifier()
```

```python
# Training classifier with above preprocess split data 
trained_model = classifier.fit(traing_input,train_label)
```

```python
# test the model 
predicted_values = trained_model.predict(testing_input)
```

```python
print(predicted_values)
```

```python
# actual values
print(test_label)
```

```python
# using accuracy score 
accuracy_score(test_label,predicted_values)
# output gonna come from 0 to 1 
# 0.66 means 66% accuracy score 
```
