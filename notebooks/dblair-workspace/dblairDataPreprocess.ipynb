```python
x = df.iloc[:,0:3].values

type(x)
x
```

```python
from sklearn.impute import SimpleImputer as si
import numpy as np

#compose model to replace missing values with mean value

dblair_impute = si(missing_values=np.nan, strategy='mean')
```

```python
#fit model
dblair_impute.fit(x[:,1:3])
x
```

```python
#apply changes from fit
x[:,1:3] = dblair_impute.transform(x[:,1:3])
x
```

```python
y = df.iloc[:,-1].values
type(y)
y
```

```python
#compose model to encode non-numerical variables with 2 categories only

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
#fit model and apply to data simultaneously
y = le.fit_transform(y)
y
```

```python
#create dummies for all variables with 3 or more categories
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
x = np.array(ct.fit_transform(x))
x
```

```python
#split data into train and test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)
```

```python
#Fit and pply model
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=12345)
```

```python
dtc_trained = dtc.fit(x_train, y_train)
```

```python
y_pred = dtc_trained.predict(x_test)
```

```python
[y_pred, y_test]
i = 0
results = []
for y in y_pred:
    if y == y_test[i]:
        results.append(True)
    else: 
        results.append(False)
    i+=1
print(results)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(f"Model Accuracy: {results.count(True)/len(results)}")
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
